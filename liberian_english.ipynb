```python
# prompt: Segment Audio into 30-second Clips

import librosa
import soundfile as sf
import os

def segment_audio(input_file, output_dir, segment_length=30):
    # Load the audio file
    y, sr = librosa.load(input_file)

    # Calculate the number of segments
    total_length = librosa.get_duration(y=y, sr=sr)
    num_segments = int(total_length // segment_length)

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Segment the audio
    for i in range(num_segments):
        start_sample = i * segment_length * sr
        end_sample = (i + 1) * segment_length * sr
        segment = y[start_sample:end_sample]
        output_file = os.path.join(output_dir, f"{os.path.splitext(os.path.basename(input_file))[0]}_segment_{i + 1}.wav")
        sf.write(output_file, segment, sr)

# Example usage
for filename in os.listdir('.'):
    if filename.endswith('.wav'):
      segment_audio(filename, 'segmented_audio')
```

```python
# prompt: Segment Script:

import yt_dlp
import os
import librosa
import soundfile as sf

# Install yt-dlp if not already installed
try:
    import yt_dlp
except ImportError:
    !pip install yt-dlp
    import yt_dlp


# Create a file named youtube_links.txt with some YouTube URLs
with open('youtube_links.txt', 'w') as f:
  f.write('https://www.youtube.com/watch?v=2yX_ntSv3J8\n') # liberia english
  f.write('https://www.youtube.com/watch?v=KHIfc-x5-cc\n') #
  # Add more YouTube links here as needed



ydl_opts = {
    'format': 'bestaudio/best',
    'outtmpl': '%(title)s.%(ext)s',
    'noplaylist': True,
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
    }]
}

with open('youtube_links.txt', 'r') as f:
    for line in f:
        url = line.strip()
        try:
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])
            print(f"Downloaded and extracted audio from {url}")
        except Exception as e:
            print(f"Error downloading or extracting audio from {url}: {e}")



def segment_audio(input_file, output_dir, segment_length=30):
    # Load the audio file
    try:
        y, sr = librosa.load(input_file)
    except Exception as e:
        print(f"Error loading audio file {input_file}: {e}")
        return

    # Calculate the number of segments
    total_length = librosa.get_duration(y=y, sr=sr)
    num_segments = int(total_length // segment_length)

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Segment the audio
    for i in range(num_segments):
        start_sample = i * segment_length * sr
        end_sample = min((i + 1) * segment_length * sr, len(y)) # Prevent index out of bounds
        segment = y[start_sample:end_sample]
        output_file = os.path.join(output_dir, f"{os.path.splitext(os.path.basename(input_file))[0]}_segment_{i + 1}.wav")
        try:
            sf.write(output_file, segment, sr)
            print(f"Created segment {output_file}")
        except Exception as e:
            print(f"Error writing segment {output_file}: {e}")

# Example usage
for filename in os.listdir('.'):
    if filename.endswith('.wav'):
        segment_audio(filename, 'segmented_audio')
```
